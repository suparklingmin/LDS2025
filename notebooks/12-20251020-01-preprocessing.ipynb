```python
import zipfile
import json
from tqdm import tqdm
import pandas as pd
```

```python
zipfilename = '../NIKL_OM_2021_v1.1_JSON.zip' # temp

utterances_list = []
speakers_list = []
```

```python
with zipfile.ZipFile(zipfilename) as z:
    for filename in tqdm(z.namelist()):
        if not filename.endswith(('.json', '.JSON')):
            continue
        
        with z.open(filename) as f:
            data = json.loads(f.read().decode('utf-8'))

            for document in data['document']:
                doc_id = document['id']

                # utterance
                for utterance in document['utterance']:
                    utterance.update(
                        {'speaker_id': f'{doc_id}_{utterance['speaker_id']}'}
                    )
                    utterances_list.append(utterance)
                
                # speaker
                for speaker in document['metadata']['speaker']:
                    speaker.update({'id': f'{doc_id}_{speaker['id']}'})
                    speakers_list.append(speaker)
```
