```python
import pandas as pd
```

```python
utterances = pd.read_csv(
    '../data/nikl/NIKL_OM_utterances.csv.tar.gz',
    compression='gzip'
)
utterances.dropna(inplace=True)
utterances = utterances[['id', 'form', 'speaker_id']]
utterances
```

```python
forms = utterances['form'][utterances['form'].str.contains(r'\b(?:완전히|완전|아주)\b')]
forms
```

```python
from collections import Counter
cnt_wanjeon = Counter(forms.str.findall(r'\b완전 (\w+)').sum())
cnt_aju = Counter(forms.str.findall(r'\b아주 (\w+)').sum())
```

```python
df = pd.DataFrame(
    [cnt_wanjeon, cnt_aju],
    index=('완전', '아주')
    ).transpose()
df.fillna(0, inplace=True)
df = df.astype('int')
df
```

```python
df.sort_values('완전', ascending=False)
```

```python
df_smooth = df + 1
df_smooth['완전/아주'] = df_smooth['완전'] / df_smooth['아주']
df_smooth.sort_values('완전/아주', ascending=True)[:30]
```

```python
print((df['완전'] > 0).sum()) # type-freq
print(df['완전'].sum()) # token-freq
```

```python
def get_ttr(column):
    token_freq = column.sum()
    type_freq = (column > 0).sum()

    return type_freq / token_freq
```

```python
df.apply(get_ttr) # 다양성 비교
```
