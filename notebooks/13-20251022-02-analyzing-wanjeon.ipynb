```python
import pandas as pd
```

```python
utterances = pd.read_csv(
    '../data/nikl/NIKL_OM_utterances.csv.tar.gz',
    compression='gzip'
    )
utterances.dropna(inplace=True)
utterances = utterances[['id', 'form', 'speaker_id']]
utterances
```

```python
speakers = pd.read_csv(
    '../data/nikl/NIKL_OM_speakers.csv.tar.gz',
    compression='gzip'
    )
speakers.dropna(inplace=True)
speakers = speakers[['id', 'age', 'sex']]
speakers.rename(columns={'id': 'speaker_id'}, inplace=True)
speakers
```

```python
df = utterances.merge(speakers, on="speaker_id")
df
```

```python
queries = ('완전', '아주')
```

```python
def count_queries(df, queries=queries):
    output = {} # {'아주': 3, '완전': 7}
    for query in queries:
        output[query] = df['form'].str.count(fr'\b{query}\b').sum()
    
    return pd.Series(output)
```

```python
counts = count_queries(df, queries)
counts 
```

```python
def get_prop(series):
    return series / series.sum()
```

```python
get_prop(counts)
```

```python
counts_by_age = df.groupby(['age']).apply(count_queries)
counts_by_age
```

```python
counts_by_age.apply(get_prop, axis=1)
```

```python
counts_by_sex = df.groupby(['sex']).apply(count_queries)
counts_by_sex
```

```python
counts_by_sex.apply(get_prop, axis=1)
```

```python
counts_by_age_sex = df.groupby(['age', 'sex']).apply(count_queries)
counts_by_age_sex
```

```python
counts_by_age_sex.apply(get_prop, axis=1)
```
